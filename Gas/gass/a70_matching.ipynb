```python
gass = Table.read('Gas/gass/gass.ascii',format='ascii')
alfalfa = Table.read('../fits/ALFALFA/a70.csv')
```

```python
def redshift_from_dist(D):
    return (70/3e5)*D

alfalfa['redshift'] = redshift_from_dist(alfalfa['dist'])
```

```python
gass_ra,gass_dec = [gass['RA'],gass['DEC']]
alfa_ra,alfa_dec = [alfalfa['OCra'],alfalfa['OCdec']]

_ = plt.scatter(alfa_ra,alfa_dec,s=2,label='$\alfa70$')
_ = plt.scatter(gass_ra,gass_dec,color='r',s=5,label='GASS')
```

```python
gass_radec = ICRS(gass_ra,gass_dec,unit=(u.degree,u.degree))
a70_radec = ICRS(alfa_ra,alfa_dec,unit=(u.degree,u.degree))
i1,d1,_ = match_coordinates_sky(a70_radec,gass_radec)
# Plot up the distances:
d1_arcsec = d1*3600
_ = plt.hist(d1_arcsec,bins=np.linspace(0,20,101),histtype='stepfilled',color='b',alpha=0.5)
plt.xlabel('$\Delta_{\mathrm{alfa-GASS}} (arcsec)$')
plt.ylabel('$N_{\mathrm{gal}}$')

i0 = range(len(i1))
matching_array = np.array([alfalfa['OCra'],alfalfa['OCdec'],i0,i1,d1_arcsec]).T
matching_table = Table(matching_array,names=('ra','dec','i_alfa','i_gass','d'))
keep_matches = matching_table['d'] <= 5 # within 5"?
matched_table = matching_table[keep_matches]
print('Have {} matches of GASS w.r.t. ALFALFA'.format(len(matched_table)))
```

```python
matched_gass = gass[matched_table['i_gass'].astype(int)]
matched_alfa = alfalfa[matched_table['i_alfa'].astype(int)]
```

```python
print(matched_alfa.colnames,'\n' ,matched_gass.colnames)
```

```python
#_ = plt.hist(np.abs(matched_gass['zSDSS']-matched_alfa['redshift']),alpha=0.4,histtype='stepfilled',bins=100)
z_ok = np.abs(matched_gass['zSDSS']-matched_alfa['redshift']) <= 0.002

matched_alfa = matched_alfa[z_ok]
matched_gass = matched_gass[z_ok]

_ = plt.scatter(matched_alfa['loghimass'],matched_gass['logMHI'])
plt.axis([9,11,9,11])
plt.xlabel(r'$log(M_{HI})$ ($\alpha$70)')
plt.ylabel(r'$log(M_{HI})$ (GASS)')
```
