## First, look at HI w. ALFALFA:

```python
gz_data = Table.read('../matched_fits/gz_data.fits')
a70_data = Table.read('../../../fits/ALFALFA/a70.csv')
```

```python
import astropy.units as u
import astropy.constants as const

def z_from_dist(D):
    # Convert distance (in Mpc) -> z
    D = D*u.Mpc
    c = const.c.to(u.km/u.s)
    H_0 = 70*(u.km/u.s/u.Mpc)
    z = (H_0/c)*D
    return z


def dist_from_z(z):
    # Convert z -> distance (in Mpc)
    c = const.c.to(u.km/u.s)
    H_0 = 70*(u.km/u.s/u.Mpc)
    D = (c/H_0)*z
    return D

a70_data['redshift'] = z_from_dist(a70_data['dist'])
```

```python
ok_detection = a70_data['detcode'] == 1 # high SNR detection
a70_ok = a70_data[ok_detection]
```

```python
from combining_catalogues import match_sky_restricted, make_matched_catalogue

match_ids = match_sky_restricted(gz_data,a70_ok,max_separation=10,max_dz=0.01,
                                 reference_xyz=['ra_1','dec_1','REDSHIFT_1'],
                                 match_xyz=['OCra','OCdec','redshift'])

a70_matched = make_matched_catalogue(gz_data,a70_ok,match_ids)
```

```python
samples = Table.read('../matched_fits/samples.fits')
print(np.sum(samples['in_redshift']*a70_matched['mask']))
```

```python
has_a70 = a70_matched['mask']

def plot_footprint_region(axis_limits):

    _ = plt.scatter(gz_data['ra_1'],gz_data['dec_1'],alpha=0.1)
    _ = plt.scatter(gz_data['ra_1'][has_a70],gz_data['dec_1'][has_a70],color='r')
    plt.axis(axis_limits)
    return None

fig, axarr = plt.subplots(3,3,figsize=(15,15))
axarr = axarr.ravel()

plt.sca(axarr[0])
plot_footprint_region((0,360,-10,90))
plt.sca(axarr[1])
plot_footprint_region((25,50,-5,10))
plt.sca(axarr[2])
plot_footprint_region((320,340,-5,10))
plt.sca(axarr[3])
plot_footprint_region((100,250,-5,10))
plt.sca(axarr[4])
plot_footprint_region((100,250,15,25))
plt.sca(axarr[5])
plot_footprint_region((100,280,20,35))
plt.sca(axarr[6])
plot_footprint_region((140,230,0,18))
```

```python
from astropy.coordinates import match_coordinates_sky, SkyCoord
import astropy.units as u

a70_in_region = ((a70_data['OCra'] > 140) &
                 (a70_data['OCra'] < 230) &
                 (a70_data['OCdec'] > 0) &
                 (a70_data['OCdec'] < 18) &
                 (a70_data['redshift'] < 0.05))
                 
print('{} HI detections in the defined region'.format(np.sum(a70_in_region)))

gz_in_region = ((gz_data['RA'] > 140) &
                (gz_data['RA'] < 230) &
                (gz_data['DEC'] > 0) &
                (gz_data['DEC'] < 18) &
                (gz_data['REDSHIFT_1'] < 0.05))

print('{} GZ galaxies in the defined region'.format(np.sum(gz_in_region)))

gz_coord = SkyCoord(ra=gz_data['RA'][gz_in_region]*u.degree, 
                     dec=gz_data['DEC'][gz_in_region]*u.degree) 
a70_coord = SkyCoord(ra=a70_data['OCra'][a70_in_region]*u.degree, 
                     dec=a70_data['OCdec'][a70_in_region]*u.degree) 

idx, sep2d, _ = match_coordinates_sky(gz_coord,a70_coord)

_ = plt.hist(sep2d.value,bins=np.linspace(0,10,100),color='k',histtype='stepfilled')
# 'in_region' = sep2d < 0.4?

gz_coord = SkyCoord(ra=gz_data['RA']*u.degree, 
                     dec=gz_data['DEC']*u.degree) 
a70_coord = SkyCoord(ra=a70_data['OCra']*u.degree, 
                     dec=a70_data['OCdec']*u.degree) 

idx, sep2d, _ = match_coordinates_sky(gz_coord,a70_coord)

_ = plt.hist(sep2d.value,bins=np.linspace(0,10,100),color='r',histtype='step',lw=2)

in_survey_region = (gz_data['REDSHIFT_1'] < 0.05) & (sep2d.value < 1)
```

```python
fig, axarr = plt.subplots(1,2,figsize=(10,5))
in_redshift = gz_data['REDSHIFT_1'] < 0.05

axarr[0].scatter(gz_data['RA'][in_redshift],gz_data['DEC'][in_redshift],color='b',alpha=1)
axarr[1].scatter(gz_data['RA'][in_redshift],gz_data['DEC'][in_redshift],color='b',alpha=1)
axarr[0].scatter(a70_data['OCra'],a70_data['OCdec'],color='k',alpha=0.2)
axarr[1].scatter(gz_data['RA'][in_survey_region == True],
                 gz_data['DEC'][in_survey_region == True],color='r',alpha=1)
axarr[0].set_title('ALFALFA70 region')
axarr[1].set_title('SDSS matched survey region')
#plt.axis((150,170,10,15))
```

```python
from bar_plots import comparison_bar_chart

samples = Table.read('../matched_fits/samples.fits')

spirals = (samples['in_mass_limit']) & (samples['axial_ok']) & (samples['in_spiral'])
targeted = (spirals) & (in_survey_region)

targeted_gals = (targeted) & (spirals)

detection_table = Table()
labels = ['1','2','3','4','5+']

for m in range(5):
    in_m = (targeted_gals) & (samples['m'] == m)
    has_a70 = a70_matched['mask'][in_m]
    N_total = len(has_a70)
    N_gas = np.sum(has_a70)
    detection_table[labels[m]] = [N_gas,N_total-N_gas]

_ = comparison_bar_chart(detection_table,labels=[r'$\alpha 70$ detection','No detection'],
                         colors='rw',normalise='percent',linewidth=1,ylabel='$\%$ detected')

print(detection_table)#, '\n', 

#plt.savefig('../output_figures/a70_detections.pdf')
```

```python
coldgass = Table.read('../../../fits/ALFALFA/COLDGASS_DR3.fits')

match_ids = match_sky_restricted(gz_data,coldgass,max_separation=10,max_dz=0.001,
                                 reference_xyz=['ra_1','dec_1','REDSHIFT_1'],
                                 match_xyz=['RA','DEC','Z_SDSS'])

coldgass_matched = make_matched_catalogue(gz_data,coldgass,match_ids)
```

```python
# look in to the detection fractions vs. arm number?
samples = Table.read('../matched_fits/samples.fits')

in_spiral = (samples['axial_ok']) & (samples['in_spiral'])
in_a70 = a70_matched['mask'] == True
in_coldgass = coldgass_matched['mask'] == True
spiral_and_coldgass = (in_spiral) & (in_coldgass) & (samples['in_mass_limit'])
```

```python
# M*-limited sample?
in_mstar = (in_spiral) & (in_a70) & (samples['in_mass_limit'])
# Vmax sample?
in_vmax = (in_spiral) & (in_a70) & (samples['in_vmax_limit'])
print('M*-limited: {} spirals w. HI'.format(np.sum(in_mstar)))
print('Vmax-limited: {} spirals w. HI'.format(np.sum(in_vmax)))

print('Cross-match w. COLDGASS: {} spirals'.format(np.sum(spiral_and_coldgass)))
```

```python
# save the outputs:
a70_matched.write('../matched_fits/a70_matched.fits',overwrite=True)
coldgass_matched.write('../matched_fits/coldgass_matched.fits',overwrite=True)
```

```python
from bar_plots import comparison_bar_chart

in_survey_area = (gz_data['dec_1'] >= -0.2) & (gz_data['dec_1'] <= 30.2)

in_mass_limit = samples['in_mass_limit']
in_a70 = a70_matched['mask']
a70_detections = Table()
column_labels = ['$m=1$','$m=2$','$m=3$','$m=4$','$m=5+$']
for m in range(5):
    in_m = samples['m'] == m
    detected = (in_m) & (in_mass_limit) & (in_a70) & (in_spiral) & (in_survey_area)
    non_detected = (in_m) & (in_mass_limit) & (in_a70 == False) & (in_spiral) & (in_survey_area)
    a70_detections[column_labels[m]] = [np.sum(detected),np.sum(non_detected)]
    
print(a70_detections)
```

```python
coldgass_detections = Table()
column_labels = ['$m=2$','$m>2$']
m_coldgass[m == 1] = 1
m_coldgass[m > 1] = 2

column_totals = []

for i, m in enumerate([1,2]):
    in_m = m_coldgass == m
    coldgass_m = coldgass_spirals[in_m]
    detected_H2 = coldgass_m['FLAGCO'] == 1
    non_detected_H2 = coldgass_m['FLAGCO'] == 2
    detected_HI = coldgass_m['FLAGHI'] != 4
    non_detected_HI = coldgass_m['FLAGHI'] == 4
    coldgass_detections[column_labels[i]] = [np.sum(detected_H2),np.sum(non_detected_H2),
                                             np.sum(detected_HI),np.sum(non_detected_HI)]
    column_totals.append(np.sum(in_m))
    
labels = ['CO detected','CO non detected','HI detected','HI non detected']
_ = comparison_bar_chart(coldgass_detections,labels,'bwrw',normalise='percent',ylabel='$\%$',linewidth=2,
                         column_totals=column_totals)

#plt.title('ALFALFA HI detections')
```
