## What is the fraction of the _total_ galaxy population with each arm number?

```python
# Preamble

from astropy.table import Table, column
import numpy as np
import os

%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
# better-looking plots
plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.figsize'] = (10.0, 8)
plt.rcParams['font.size'] = 18
mpl.ticker.AutoLocator.default_params['nbins'] = 5
mpl.ticker.AutoLocator.default_params['prune'] = 'both'

mpl.rcParams['ps.useafm'] = True
mpl.rcParams['pdf.use14corefonts'] = True
mpl.rcParams['text.usetex'] = True
```

```python
import sys
sys.path.append('../../../astro_codes/')
#------------------------------------
```

```python
samples = Table.read('../matched_fits/samples.fits')
nsa_data = Table.read('../matched_fits/nsa_matched.fits')
```

```python
in_sample = samples['in_mass_limit']
in_spiral = (samples['in_spiral']) & (samples['axial_ok'])
```

```python
from scipy.stats import binned_statistic

N_bins = 20
z_bins = np.linspace(0.02,0.055,N_bins+1)

def total_bin_numbers(z_bins,sample):

    z_mean, _, bin_assignments = binned_statistic(sample,sample,
                                                  bins=z_bins)

    z_bin_centres = [z_bins[i] + (z_bins[i+1]-z_bins[i])/2 for i in range(N_bins)]
    N_z_bin = []
    for i in range(N_bins):
        N_z_bin.append(np.sum(bin_assignments == i+1))
        
    return z_bin_centres, N_z_bin
```

```python
sample = nsa_data['Z'][in_sample]

z_bin_centres, N_total = total_bin_numbers(z_bins,sample)
_, N_1 = total_bin_numbers(z_bins,nsa_data[(in_sample) & (in_spiral) & (samples['m'] == 0)]['Z'])
_, N_2 = total_bin_numbers(z_bins,nsa_data[(in_sample) & (in_spiral) & (samples['m'] == 1)]['Z'])
_, N_3 = total_bin_numbers(z_bins,nsa_data[(in_sample) & (in_spiral) & (samples['m'] == 2)]['Z'])
_, N_4 = total_bin_numbers(z_bins,nsa_data[(in_sample) & (in_spiral) & (samples['m'] == 3)]['Z'])
_, N_5 = total_bin_numbers(z_bins,nsa_data[(in_sample) & (in_spiral) & (samples['m'] == 4)]['Z'])

_, N_spiral = total_bin_numbers(z_bins,nsa_data[(in_sample) & (in_spiral)]['Z'])
```

```python
f_1 = np.array(N_1)/np.array(N_total)
f_2 = np.array(N_2)/np.array(N_total)
f_3 = np.array(N_3)/np.array(N_total)
f_4 = np.array(N_4)/np.array(N_total)
f_5 = np.array(N_5)/np.array(N_total)

f_spiral = np.array(N_spiral)/np.array(N_total)
```

```python
fig, axarr = plt.subplots(5,1,figsize=(10,15))
plt.subplots_adjust(hspace=0,wspace=0)

axarr[0].plot(z_bin_centres,f_1)
axarr[1].plot(z_bin_centres,f_2)
axarr[2].plot(z_bin_centres,f_3)
axarr[3].plot(z_bin_centres,f_4)
axarr[4].plot(z_bin_centres,f_5)
```

```python
plt.plot(z_bin_centres,f_spiral)

_ = plt.ylim((0,1))
```
