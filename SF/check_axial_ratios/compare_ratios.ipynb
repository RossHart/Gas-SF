```python
import astropy.units as u

def axial_ratio_to_inclination(axial_ratio,q=0.2,unit='degree'):
    cos2_i = (axial_ratio-q**2)/(1-q**2)
    cos_i = np.sqrt(cos2_i)
    i = np.arccos(cos_i)*u.radian
    if unit == 'degree':
        i = i.to(u.degree)
    return i.value

def ba_to_i(axial_ratio,q):
    cos2_i = (axial_ratio-q**2)/(1-q**2)
    cos_i = np.sqrt(cos2_i)
    i = np.arccos(cos_i)*u.radian
    i = i.to(u.degree)
    return i.value
    

dr7['isoAB'] = dr7['isoB_g']/dr7['isoA_g']
dr7['isoAB_err'] = dr7['isoAB']*(((dr7['isoAErr_g']/dr7['isoA_g'])**2 +
                                 (dr7['isoBErr_g']/dr7['isoB_g'])**2)**(0.5))

dr7['i'] = axial_ratio_to_inclination(dr7['isoAB'])
```
