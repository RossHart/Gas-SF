```python
from astroML.plotting import hist

def histogram_data(data,reference_data,data_range=None,bins=None,
                   line_properties=None,fill_properties=None):
    
    fp = {'color':'k',
          'alpha':0.3}  

    lp = {'color':'k',
          'alpha':1,
          'linewidth':1,
          'linestyle':'solid'}
    
    if line_properties != None:
        for l in line_properties.keys():
            lp[l] = line_properties[l]
    if fill_properties != None:
        for f in fill_properties.keys():
            fp[f] = fill_properties[f]
    if data_range != None:
        data,reference_data = [u[(u >= data_range[0]) & (u <= data_range[1])]
                               for u in [data,reference_data]]
    
    if bins == None:
        bins = 'knuth'
            
    _ = hist(reference_data,bins=bins,normed=True,histtype='stepfilled',
             color=fp['color'],alpha=fp['alpha'])
    
    bins = _[1]
    _ = hist(data,bins=bins,normed=True,histtype='step',
             color=lp['color'],alpha=lp['alpha'],
             linewidth=lp['linewidth'],linestyle=lp['linestyle'])
    
    return bins
```

```python
has_inclination = ((ba_spirals['ba_isophote_g'] >= 0) 
                   & (ba_spirals['ba_isophote_g'] <= 1))

inclinations = ba_spirals[has_inclination]['ba_isophote_g']
m_inclinations = gz_spirals[has_inclination]['m']
```

```python
from scipy.stats import binned_statistic
import scipy.stats.distributions as dist

def get_fractional_errors(k,n,c=0.683):
    
    p_lower = dist.beta.ppf((1-c)/2.,k+1,n-k+1)
    p_upper = dist.beta.ppf(1-(1-c)/2.,k+1,n-k+1)
    
    return p_lower,p_upper
```

```python
inc, _, bins = binned_statistic(inclinations,inclinations,bins=np.linspace(0.3,1,21),
                                statistic='median')

inc = inc[np.isfinite(inc)]

f_array = np.zeros((len(inc),3))

for i, b in enumerate(inc):
    m_bin = m_inclinations[bins == i+1]
    k, n = [np.sum(m_bin == 1),len(m_bin)]
    f = k/n
    f_low, f_high = get_fractional_errors(k,n)
    f_array[i,:] = [f_low, f, f_high]

f_table = Table(f_array,names=('f_low','f','f_high'))
_ = plt.plot(inc,f_table['f'],color=colors_long[m],linewidth=2)
_ = plt.fill_between(inc,f_table['f_low'],f_table['f_high']
                         ,color=colors_long[m],alpha=0.4)
    
plt.xlabel(r'$\mathrm{(b/a)}_g$')
plt.ylabel(r'$f_{m=2}$')
```

```python
import astropy.units as u

def axial_ratio_to_inclination(axial_ratio,q=0.2,unit='degree'):
    cos2_i = (axial_ratio-q**2)/(1-q**2)
    cos_i = np.sqrt(cos2_i)
    i = np.arccos(cos_i)*u.radian
    if unit == 'degree':
        i = i.to(u.degree)
    return i.value


def inclination_to_axial_ratio(inclination,q=0.2,unit='degree'):
    if unit == 'degree':
        inclination = inclination*u.degree
        inclination = inclination.to(u.radian)
        cos2_i = np.cos(inclination)**2
    else:
        inclination = inclination*u.radian
    axial_ratio = (1-q**2)*(cos2_i) + q**2
    return axial_ratio
```
