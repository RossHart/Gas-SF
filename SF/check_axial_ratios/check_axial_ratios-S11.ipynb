```python
from astroML.plotting import hist

def histogram_data(data,reference_data,data_range=None,bins=None,
                   line_properties=None,fill_properties=None):
    
    fp = {'color':'k',
          'alpha':0.3}  

    lp = {'color':'k',
          'alpha':1,
          'linewidth':1,
          'linestyle':'solid'}
    
    if line_properties != None:
        for l in line_properties.keys():
            lp[l] = line_properties[l]
    if fill_properties != None:
        for f in fill_properties.keys():
            fp[f] = fill_properties[f]
    if data_range != None:
        data,reference_data = [u[(u >= data_range[0]) & (u <= data_range[1])]
                               for u in [data,reference_data]]
    
    if bins == None:
        bins = 'knuth'
            
    _ = hist(reference_data,bins=bins,normed=True,histtype='stepfilled',
             color=fp['color'],alpha=fp['alpha'])
    
    bins = _[1]
    _ = hist(data,bins=bins,normed=True,histtype='step',
             color=lp['color'],alpha=lp['alpha'],
             linewidth=lp['linewidth'],linestyle=lp['linestyle'])
    
    return bins
```

```python
from scipy.stats import binned_statistic
import scipy.stats.distributions as dist

def get_fractional_errors(k,n,c=0.683):
    
    p_lower = dist.beta.ppf((1-c)/2.,k+1,n-k+1)
    p_upper = dist.beta.ppf(1-(1-c)/2.,k+1,n-k+1)
    
    return p_lower,p_upper
```

```python
has_inclination = ba_spirals['i']/ba_spirals['e_i'] >= 3
bulge_ok = (ba_spirals['__B_T_r'] <= 0.5) & (ba_spirals['__B_T_r']/ba_spirals['e__B_T_r'] > (1/3))
inclination_mask = (has_inclination) & (bulge_ok)

print('{}/{} ({}%) have measured *good* inclinations'.format(np.sum(inclination_mask),len(inclination_mask),
                                                             np.round(np.sum(inclination_mask)/len(inclination_mask)*100
                                                                      ,decimals=2)))
```

```python
# How do the fractions of the following galaxy types change w.r.t. inclination?:

# spiral arm number?
# bar presence?
# looseness of spiral arms?

gz_inc, ba_inc = [gz_spirals[inclination_mask],ba_spirals[inclination_mask]]

f_2 = gz_inc['t11_arms_number_a32_2_debiased_rh']
f_bar = gz_inc['t03_bar_a06_bar_debiased_rh']
f_loose = gz_inc['t10_arms_winding_a30_loose_debiased_rh']
m = gz_inc['m']
inclinations = ba_inc['i']
```

```python
from matplotlib.ticker import MaxNLocator

def plot_bins(data,i_bins,bins,color='k',alpha=0.5,xlabel=None,ylabel=None):
    bin_mean = []
    bin_error = []
    for b, i in enumerate(i_bins):
        select = bins == b+1
        bin_mean.append(np.mean(data[select]))
        bin_error.append(np.std(data[select])/np.sqrt(np.sum(select)))
    bin_mean, bin_error = [np.array(u) for u in [bin_mean,bin_error]]
    
    _ = plt.plot(i_bins,bin_mean,lw=2,color=color)
    _ = plt.fill_between(i_bins,bin_mean-bin_error,bin_mean+bin_error
                         ,color=color,alpha=alpha)
    if xlabel != None:
        plt.xlabel(xlabel)
    if ylabel != None:
        plt.ylabel(ylabel)
    
    return None


def plot_bin_classifications(data,value,i_bins,bins,color='k',alpha=0.5,xlabel=None,ylabel=None):
    bin_mean = []
    bin_lower = []
    bin_upper = []
    for b, i in enumerate(i_bins):
        select = bins == b+1
        has_value = data[select] == value
        k, n = np.sum(has_value), len(has_value)
        bin_mean.append(k/n)
        p_lower, p_upper = get_fractional_errors(k,n)
        bin_lower.append(p_lower)
        bin_upper.append(p_upper)
    bin_mean, bin_lower, bin_upper = [np.array(u) for u in [bin_mean,bin_lower,bin_upper]]
    
    _ = plt.plot(i_bins,bin_mean,lw=2,color=color)
    _ = plt.fill_between(i_bins,bin_lower,bin_upper,
                         color=color,alpha=alpha)
    if xlabel != None:
        plt.xlabel(xlabel)
    if ylabel != None:
        plt.ylabel(ylabel)
    
    return None

fig, axarr = plt.subplots(1,3,figsize=(10,3.5),sharex=True,sharey=False)

i_bins, _, bins = binned_statistic(inclinations,inclinations,bins=np.linspace(0,90,20),
                                   statistic='median')

plt.sca(axarr[0])
_ = plot_bins(f_2,i_bins,bins,color='r',
              xlabel='$i$ (deg)',ylabel='$f_\mathrm{2}$')
axarr[0].yaxis.set_major_locator(MaxNLocator(6))
plt.sca(axarr[1])
_ = plot_bins(f_bar,i_bins,bins,color='purple',
              xlabel='$i$ (deg)',ylabel='$f_\mathrm{bar}$')
axarr[1].yaxis.set_major_locator(MaxNLocator(6))
plt.sca(axarr[2])
_ = plot_bins(f_loose,i_bins,bins,color='k',
              xlabel='$i$ (deg)',ylabel='$f_\mathrm{loose}$')
axarr[2].yaxis.set_major_locator(MaxNLocator(6))

axarr[0].xaxis.set_major_locator(MaxNLocator(nbins = 6, prune = 'both'))
fig.tight_layout()
```

```python
tight = f_loose <= 0.2
unbarred = f_bar <= 0.2
control = (tight) & (unbarred)

print('Control sample: {}/{} galaxies'.format(np.sum(control),len(control)))
```

```python
gz_control, inclinations_control = [u[control] for u in [gz_inc, ba_inc]]

fig, axarr = plt.subplots(1,2,figsize=(10,5))

i_bins, _, bins = binned_statistic(inclinations_control['i'],inclinations_control['i']
                                   ,bins=np.linspace(0,90,15),statistic='median')

plt.sca(axarr[0])
_ = plot_bins(gz_control['t11_arms_number_a32_2_debiased_rh'],i_bins,bins,
              color='r',xlabel='$i$ (deg)',ylabel ='$f_\mathrm{2}$')
plt.sca(axarr[1])
_ = plot_bin_classifications(gz_control['m'],1,i_bins,bins,
                             color='r',xlabel='$i$ (deg)',ylabel='$f_\mathrm{m=2}$')
fig.tight_layout()
axarr[0].vlines(60,0.1,0.5,color='k',linestyle='dotted',lw=3)
axarr[1].vlines(60,0.2,1,color='k',linestyle='dotted',lw=3)
```
