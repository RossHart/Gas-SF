```python
has_match = nsa['zoo_match'] != -999
nsa_matched = nsa[has_match]
zoo_matched = zoo[nsa_matched['zoo_match'].astype(int)]
```

```python
F,N,u,g,r,i,z = [nsa_matched['PETROFLUX'][:,i] for i in range(7)]
F_err,N_err,u_err,g_err,r_err,i_err,z_err = [nsa_matched['PETROFLUX_IVAR'][:,i]**(-0.5) for i in range(7)]
```

```python
filters = [F,N,u,g,r,i,z]
errors = [F_err,N_err,u_err,g_err,r_err,i_err,z_err]
names = 'FNugriz'
select = np.zeros((len(nsa_matched),7))

for f,filt in enumerate(filters):
    
    error = errors[f]
    filt_ok = np.isfinite(filt)
    error_ok = np.isfinite(error)
    within_error = filt/error >= 3 # s/n > 3
    both_ok = (filt_ok) & (error_ok) & (within_error)
    print('{}: {}/{} with S/N > 3'.format(names[f],np.sum(both_ok),len(both_ok)))
    select[:,f] = both_ok
    
filter_sn = Table(select.astype(int),names=('F','N','u','g','r','i','z'))
```

```python
# assign spiral arm numbers to galaxies?

def get_assignments(data,question_dictionary,th=None):

    answers = question_dictionary['t11_arms_number']['answers']
    
    arm_answers = ['t11_arms_number_' + A + '_debiased_rh' for A in answers[:-1]]
    arm_array = np.array([data[A] for A in arm_answers]).T
    raw_answers = ['t11_arms_number_' + A + '_weighted_fraction' for A in answers[:-1]]
    raw_array = np.array([data[A] for A in raw_answers]).T
    
    if th != None:
        select = np.max(arm_array,axis=1) > th
    else:
        select = np.ones(len(data)) == 1
    
    arm_assignments = np.argmax(arm_array,axis=1)
    raw_assignments = np.argmax(raw_array,axis=1)
    
    arm_assignments = arm_assignments[select]
    raw_assignments = raw_assignments[select]
    
    return arm_assignments,raw_assignments


def reduce_sample(full_data,questions,question,p_cut=0.5,N_cut=5,normalised_values=True):
    
    # Get the reference sample from the previous data:
    
    previous_q = questions[question]['pre_questions']
    previous_a = questions[question]['pre_answers']
    
    if normalised_values == True:
        suffix = '_debiased_rh_normalised'
    else:
        suffix = '_debiased_rh'
    
    if previous_q != None:
        
        p_col = np.ones(len(full_data))
        
        for m in range(len(previous_q)):
            p_col = p_col*(full_data[previous_q[m] + '_' + previous_a[m] + suffix])
        N_col = (full_data[previous_q[-1] + '_' + previous_a[-1] + '_count'])
        
        select = (p_col > p_cut) & (N_col >= N_cut)
        data_reduced = full_data[select]
        print('{}/{} galaxies with p>{} and N>={}.'.format(len(data_reduced),
                                                          len(full_data),p_cut,N_cut))
    
    else:
        data_reduced = full_data.copy()
        select = np.ones(len(full_data)) == 1
        print('Primary question, so all {} galaxies used.'.format(len(data_reduced)))
    
    return select

#--------------------------------------
os.chdir('../Debiasing/')
import make_dictionaries
questions = make_dictionaries.questions
os.chdir('../star_formation/')

arm_assignments,_ = get_assignments(zoo_matched,questions)

spiral = reduce_sample(zoo_matched,questions,'t11_arms_number')
remove_ct = (zoo_matched['t04_spiral_a08_spiral_count'] 
             - zoo_matched['t11_arms_number_a37_cant_tell_count'] >= 5)

in_spiral = (spiral) & (remove_ct)
in_lum_limit = zoo_matched['PETROMAG_MR'] <= -20.03
in_mass_limit = zoo_matched['LOGMSTAR_BALDRY06'] >= 9.8
mass_limited_spirals = in_spiral*in_lum_limit*in_mass_limit

print('---> {} spiral galaxies in M* limited sample'.format((mass_limited_spirals).sum()))
```

```python
# NUV-u colour histograms?

spiral_data = nsa_matched[mass_limited_spirals]
spiral_arm_m = arm_assignments[mass_limited_spirals]
```

```python
def mag_to_luminosity(Mag,M_sun):
    return 10**(1/2.5*(M_sun-Mag))
```

```python
fig,axarr = plt.subplots(5,1,figsize=(10,15),sharex=True,sharey=True)
plt.subplots_adjust(hspace=0,wspace=0)

colors_long = ['orange','r','m','g','b']
B = np.linspace(-20,-15,21)

for m in range(5):
    
    NUV = spiral_data['ABSMAG'][:,1]
    _ = axarr[m].hist(NUV,histtype='stepfilled',color='k',alpha=0.4,normed=True,bins=B)
    NUV_m = NUV[spiral_arm_m == m]
    _ = axarr[m].hist(NUV_m,histtype='step',color=colors_long[m],lw=3,normed=True,bins=B)
    
    axarr[m].set_ylabel('normalised \n density')
    
axarr[-1].set_xlabel('$M_{NUV}$')
```

```python
# now luminosity:

fig,axarr = plt.subplots(5,1,figsize=(10,15),sharex=True,sharey=True)
plt.subplots_adjust(hspace=0,wspace=0)

B = np.linspace(10,12,21)

for m in range(5):
    
    NUV = mag_to_luminosity(spiral_data['ABSMAG'][:,1],10.18)
    _ = axarr[m].hist(np.log10(NUV),histtype='stepfilled',color='k',alpha=0.4,normed=True,bins=B)
    NUV_m = NUV[spiral_arm_m == m]
    _ = axarr[m].hist(np.log10(NUV_m),histtype='step',color=colors_long[m],lw=3,normed=True,bins=B)
    
    axarr[m].set_ylabel('normalised \n density')
    
axarr[-1].set_xlabel('$L_{NUV}(L_{\odot})$')
axarr[-1].set_ylim(0,2.5)
```
