```python
samples = Table.read('../matched_fits/samples.fits')
masks = Table.read('../matched_fits/bad_flux_mask.fits')
sfr_data = Table.read('../matched_fits/sfr_table.fits')
chang_out_data = Table.read('../matched_fits/chang_out_matched.fits')
nsa_data = Table.read('../matched_fits/nsa_matched.fits')

print('Loaded the data')
```

```python
# first, define the star-forming sample:

in_sf = (samples['in_vmax_limit']*samples['star_forming']).astype(bool)
ok_flux = (masks['fuv']*masks['w4']).astype(bool)
reference_sample = (in_sf*ok_flux).astype(bool)

print('{} galaxies in vmax, with {} star-forming, and {} with reliable fluxes'.format(np.sum(samples['in_vmax_limit']),
                                                                                      np.sum(in_sf),
                                                                                      np.sum(reference_sample)))

mass = chang_out_data['lmass50_all']
ssfr = sfr_data['total'] - mass
vmax_weight = samples['mass_weighting']
```

```python
def irx_beta_b12(beta):
    irx = np.log10((10**(0.396*(beta+2.046))-1)/0.373)
    return irx


def irx_beta_k04(beta):
    irx = np.log10(10**(2.1+0.85*beta)-0.95)
    return irx
```

```python
beta_guide = np.linspace(-3,2,1000)
irx_guide_b12 = irx_beta_b12(beta_guide)
irx_guide_k04 = irx_beta_k04(beta_guide)

_ = plt.plot(beta_guide,irx_guide_b12)
_ = plt.plot(beta_guide,irx_guide_k04)
```

```python
from luminosities_magnitudes_and_distances import solar_to_lum, lum_to_Mag, Mag_to_lum

ok_fluxes = (masks['fuv']) & (masks['nuv']) & (masks['w4'])

beta = ((nsa_data['M_corrected'][:,0] - nsa_data['M_corrected'][:,1]))/(2.5*math.log10(2315.66/1538.62)) - 2

l_fuv = Mag_to_lum(nsa_data['M_corrected'][:,0],1538.62)
l_w4 = solar_to_lum(chang_out_data['lrest_w4'])
l_tir_estimated = 1.445 + 0.945*chang_out_data['lrest_w4']
l_tir = chang_out_data['ldust50_all']
l_tir = solar_to_lum(l_tir)
l_tir_estimated = solar_to_lum(l_tir_estimated)
irx = np.log10(l_tir/l_fuv)
irx_estimated = np.log10(l_tir_estimated/l_fuv)

_ = plt.scatter(beta[ok_fluxes], irx[ok_fluxes],alpha=0.2,color='k')
plt.plot(beta_guide,irx_guide_k04,label='Kong+04')
plt.plot(beta_guide,irx_guide_b12,label='Boquien+12')
plt.legend()

plt.axis((-2.5,1,-1,2))
plt.xlabel(r'$\beta$')
plt.ylabel('IRX')
```

```python
from contours import kde_contour

fig, axarr = plt.subplots(2,3,figsize=(20,13.4),sharex=True,sharey=True)
plt.subplots_adjust(hspace=0,wspace=0)
axarr = axarr.ravel()
fig.delaxes(axarr[-1])

# spiral galaxies:
in_ml_sample = ((samples['in_mass_limit']) &
                (ok_fluxes))

in_spiral_sample = ((in_ml_sample) &
                    (samples['in_spiral']) &
                    (samples['axial_ok']))

m_values = samples['m']
colors = ['orange','r','m','g','b']
figlabels = ['1','2','3','4','5+']

reference_x, reference_y = beta[in_ml_sample], irx[in_ml_sample]
xy_range = ((-2,0.5,-0.5,2))

for m in range(5):
    
    plt.sca(axarr[m])
    
    _ = kde_contour(reference_x,reference_y,xy_range=xy_range,fill=True)
    plt.plot(beta_guide,irx_guide_k04,'k--',lw=3)
    plt.plot(beta_guide,irx_guide_b12,'k:',lw=4)
    plt.axis(xy_range)
    
    in_m = (in_spiral_sample) & (m_values == m)
    x, y = beta[in_m], irx[in_m]
    line_properties = {'color':colors[m],
                       'linewidth':2}
    _ = kde_contour(x,y,xy_range=xy_range,line_properties=line_properties)
    plt.text(0.95,0.95,'m={} (N={})'.format(figlabels[m],len(x)),
             transform=axarr[m].transAxes,ha='right',va='top')
        
fig.text(0.5, 0.08, r'$\mathrm{\beta}$',
         ha='center', va='center',size=24)
fig.text(0.08, 0.5, '$\mathrm{IRX}$',
         ha='center', va='center', rotation='vertical',size=24)

plt.axis(xy_range)
#plt.savefig('../output_figures/irx_beta.pdf')
```
