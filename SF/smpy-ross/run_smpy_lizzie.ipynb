```python
# for convenience...
values = {}
values['filter'] = np.array(['F','N','u', 'g', 'r', 'i', 'z'])
values['redshift'] = synmag['z']
values['tg'] = par['tg'] / 1e9  # Gyr
values['Av'] = par['Av']
values['tau'] = par['SFH'] / 1e9  #Gyr
values['metallicity'] = np.array([float(s.split('=')[-1]) for s in par['SSPs']])
values
```

```python
def take_nearest(axisname, val, array, values=values,quiet=False):
    idx = (np.abs(values[axisname] - val)).argmin()
    if quiet == False:
        print('Found {} = {}'.format(axisname, values[axisname][idx]))
    return array.take([idx], axis=axes[axisname])
```

```python
# Example of extracting magnitudes as a function of age

#get magnitude
mags = mags_all
# restframe (redshift = 0)
mags = take_nearest('redshift', 0.0, mags)
# solar metallicity
mags = take_nearest('metallicity', 0.02, mags)
# no dust extinction
mags = take_nearest('Av', 0.0, mags)
# some SF timescale
mags = take_nearest('tau', 2.5, mags)

F, N, u, g, r, i, z = mags.squeeze()
```

```python
# for convenience...
def age_line(x, y, val, color, label=r'$\tau$', lw=3):
    plt.plot(x, y, '-', color=color, lw=lw,
             label=r'{} = {}'.format(label, val),
             zorder=1)
    plt.scatter(x, y, c=values['tg'], s=50, zorder=2)
    
    print(values['tg'].shape,x.shape,y.shape)
```

```python
# plot the track of an aging galaxy in restframe colour-colour space
age_line(u-r,r-z, 2.5, 'g')
cb = plt.colorbar()
cb.ax.set_ylabel('age [Gyr]')
plt.legend(loc = 'upper left', fontsize='small')
plt.xlabel('$NUV-u$')
_ = plt.ylabel('$u-r$')
```

```python
# for convenience...
def color_color_age_line_par(par, val, mags, linecolor='k',
                             label=r'$\tau$',
                             values=values,
                             colxb='u', colxr='r',
                             colyb='r', colyr='z'):
    mags = take_nearest(par, val, mags)
    f = values['filter']
    colx = mags[f == colxb] - mags[f == colxr]
    coly = mags[f == colyb] - mags[f == colyr]
    colx = colx.squeeze()
    coly = coly.squeeze()
    age_line(colx, coly, val, linecolor, label)
```

```python
def mag_to_lum(M):
    return 10**(-M/2.5)

def lum_to_mag(L):
    return -2.5*np.log10(L)
    

def get_population(mags_all,t,tau,Av=0,metallicity=0.02,luminosity=True):
    
    mags = take_nearest('redshift',0,mags_all,quiet=True)
    mags = take_nearest('metallicity',metallicity,mags,quiet=True)
    mags = take_nearest('Av',Av,mags,quiet=True)
    mags = take_nearest('tg',t,mags,quiet=True)
    mags = take_nearest('tau',tau,mags,quiet=True).squeeze()
    
    if luminosity:
        lum = mag_to_lum(mags)
        return lum
    else:
        return mags
    
'''
    old_lum = mag_to_lum(old_mags)
    young_lum = mag_to_lum(young_mags)
    
    # now combine:
    combined_lum = frac_young*young_lum + (1-frac_young)*old_lum
    combined_mags = lum_to_mag(combined_lum)
    
    return combined_mags
'''
```

```python
def scatter_old_young(mags_all,ttau_young,ttau_old,f1='u',f2='r',f3='r',f4='z',color='k'
                      ,fracs=np.linspace(0,1,10)):
    
    t_young,tau_young = ttau_young
    t_old,tau_old = ttau_old
    
    young_lum = get_population(mags_all,t_young,tau_young)
    old_lum = get_population(mags_all,t_old,tau_old)

    mag_array = np.zeros((len(fracs),8))

    for i,frac in enumerate(fracs):
        lum_total = frac*young_lum + (1-frac)*old_lum
        mag_total = lum_to_mag(lum_total)
        mag_array[i,1:] = mag_total
        mag_array[i,0] = frac
    mag_table = Table(mag_array,names='fFNugriz')
    
    _ = plt.plot(mag_table[f1]-mag_table[f2],mag_table[f3]-mag_table[f4],color=color,lw=2,
                 label='$\tau_{old}$'+'={}'.format(tau_old)+'$\tau_{young}$'+'={}'.format(tau_young))
    _ = plt.scatter(mag_table[f1]-mag_table[f2],mag_table[f3]-mag_table[f4],c=1-mag_table['f'],
                    s=30)
    
    return None
```
