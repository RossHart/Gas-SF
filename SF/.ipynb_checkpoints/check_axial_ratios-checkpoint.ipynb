```python
from astropy.table import Table
import numpy as np
import os

%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
# better-looking plots
plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.figsize'] = (10.0, 8)
plt.rcParams['font.size'] = 18
mpl.ticker.AutoLocator.default_params['nbins'] = 5
mpl.ticker.AutoLocator.default_params['prune'] = 'both'

mpl.rcParams['ps.useafm'] = True
mpl.rcParams['pdf.use14corefonts'] = True
mpl.rcParams['text.usetex'] = True
```

```python
gz_data = Table.read('../../fits/full_sample_debiased_w_low_z.fits')
```

```python
ba_data = Table.read('../../fits/Axial_ratios/matched_axial_ratios.fits')
```

```python
os.chdir('../../astro_codes/')
from volume_limiting import get_volume_limited_sample
from mass_limiting import get_mass_limit,baldry_equation
from bar_plots import comparison_bar_chart
import make_dictionaries
questions = make_dictionaries.questions
os.chdir('../Gas-SF/SF/')
```

```python
data = get_volume_limited_sample(gz_data,0.055,17.0,column_names=['REDSHIFT_1','PETROMAG_MR'],
                                 low_z_limit=0.02,append_column=True)

print('volume-limited sample: {}/{} galaxies'.format(np.sum(data['in_volume_limit']), len(data)))
```

```python
p_spiral = (data['t01_smooth_or_features_a02_features_or_disk_debiased_rh']*
            data['t02_edgeon_a05_no_debiased_rh']*
            data['t04_spiral_a08_spiral_debiased_rh'])

N_spiral = (data['t04_spiral_a08_spiral_count'] - 
            data['t11_arms_number_a37_cant_tell_count'])

p_ok = p_spiral > 0.5
N_ok = N_spiral >= 5
in_spiral = (p_ok) & (N_ok)
data['in_spiral'] = in_spiral

# assign arm numbers:

question = 't11_arms_number'
answers = questions[question]['answers'][:-1]
answer_array = np.array([data[question + '_' + A + '_debiased_rh'].data for A in answers]).T
m = np.argmax(answer_array,axis=1)
data['m'] = m
```

```python
select_spirals = (data['in_spiral']) & (data['in_volume_limit'])
ba_spirals = ba_data[select_spirals]
gz_spirals = gz_data[select_spirals]
```

```python
fig,axarr = plt.subplots(5,1,figsize=(10,15))
plt.subplots_adjust(hspace=0,wspace=0)

B = np.linspace(0,1,26)
colors_long = ['orange','r','m','g','b']

for m in range(5):
    
    plt.sca(axarr[m])
    
    select_m = gz_spirals['m'] == m
    ba_m = ba_spirals[select_m]
    _ = plt.hist(ba_spirals['ba_isophote_g'],bins=B,color='k',histtype='stepfilled',alpha=0.5,normed=True)
    _ = plt.hist(ba_m['ba_isophote_g'],bins=B,lw=3,color=colors_long[m],normed=True,histtype='step')
    plt.ylabel('normalised \n density')
    
axarr[-1].set_xlabel('$\mathrm{(b/a)}_g$')

#plt.savefig('axial_ratios.pdf')
```

```python
import urllib
from PIL import Image

def display_image(url_name,crop_in=100):
    # Get the url name:
    urllib.request.urlretrieve(url_name,"image.jpg")
    # Open -> crop -> display -> remove the image.
    im=Image.open("image.jpg")
    l=424 # Image size
    im=im.crop((crop_in,crop_in,l-crop_in,l-crop_in))
    plt.imshow(im)
    os.remove("image.jpg")
    plt.xticks([])
    plt.yticks([])
    return None
```

```python
# get some galaxy images by inclination?
np.random.seed(0)
inclinations = np.linspace(0.2,0.9,8)

fig,axarr = plt.subplots(5,len(inclinations)-1,figsize=(14,10))
plt.subplots_adjust(hspace=0.02,wspace=0.02)

urls = gz_spirals['jpeg_url'] 

for m in range(5):
    select_m = gz_spirals['m'] == m
    ba_m = ba_spirals[select_m]

    for i,inc in enumerate(inclinations[:-1]):
        inc_range = (inc,inclinations[i+1])
        in_ba = ((ba_spirals['ba_isophote_g'] >= inc_range[0]) & 
                 (ba_spirals['ba_isophote_g'] < inc_range[1]))
        url = np.random.choice(urls[(select_m) & (in_ba)])
        plt.sca(axarr[m,i])
        display_image(url)

xlabels = ['m=1','m=2','m=3','m=4','m=5+']
for m in range(5):
    axarr[m,0].set_ylabel(xlabels[m])
for i in range(len(inclinations)-1):
    axarr[-1,i].set_xlabel(r'$\mathrm(b/a)_g \sim {}$'.format(np.round(inclinations[i]+0.05
                                                                       ,decimals=2)))

#plt.savefig('galaxy_inclinations.pdf')
```
