```python
from astropy.table import Table
import numpy as np
import os
import math
import random

%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
# better-looking plots
plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.figsize'] = (10.0, 8)
plt.rcParams['font.size'] = 18
mpl.ticker.AutoLocator.default_params['nbins'] = 5
mpl.ticker.AutoLocator.default_params['prune'] = 'both'
mpl.rcParams['ps.useafm'] = True
mpl.rcParams['pdf.use14corefonts'] = True
mpl.rcParams['text.usetex'] = True
```

```python
# first, import the GZ galaxy data:
gz_data = Table.read('../../fits/full_sample_debiased_w_low_z.fits')
```

```python
# next, import all of the matched files:
nsa_data = Table.read('matched_fits/nsa_matched.fits')
a70_data = Table.read('matched_fits/a70_matched.fits')
chang_in = Table.read('matched_fits/chang_input_matched.fits')
chang_out = Table.read('matched_fits/chang_output_matched.fits')
mpa_data = Table.read('matched_fits/mpa_sfrs_matched.fits')
mpa_i = Table.read('matched_fits/mpa_i_matched.fits')
```

# Get the spiral, lum. + M* limited samples

```python
# get spirals:
import make_dictionaries
questions = make_dictionaries.questions

in_sample = Table() # table of in volume, in mass, in spiral etc.

p_spiral = (gz_data['t01_smooth_or_features_a02_features_or_disk_debiased_rh']*
            gz_data['t02_edgeon_a05_no_debiased_rh']*
            gz_data['t04_spiral_a08_spiral_debiased_rh'])

N_spiral = (gz_data['t04_spiral_a08_spiral_count'] - 
            gz_data['t11_arms_number_a37_cant_tell_count'])

p_ok = p_spiral > 0.5
N_ok = N_spiral >= 5
in_spiral = (p_ok) & (N_ok)
in_sample['in_spiral'] = in_spiral

question = 't11_arms_number'
answers = questions[question]['answers'][:-1]
answer_array = np.array([gz_data[question + '_' + A + '_debiased_rh'].data for A in answers]).T
m = np.argmax(answer_array,axis=1)
in_sample['m'] = m

has_bar = gz_data['t03_bar_a06_bar_debiased_rh'] > gz_data['t03_bar_a07_no_bar_debiased_rh']
in_sample['bar'] =  has_bar
```

```python
from volume_limiting import get_volume_limited_sample
from mass_limiting import get_mass_limit,baldry_equation

in_volume = get_volume_limited_sample(gz_data,0.055,17,['REDSHIFT_1','PETROMAG_MR'],0.02,False)

mass_limit = get_mass_limit(gz_data,0.055,17,baldry_equation,0.02,column_names=['REDSHIFT_1','PETROMAG_MR']
                            ,colours=['PETROMAG_MU','PETROMAG_MR'],Mag_sun=4.67)

in_mass = gz_data['LOGMSTAR_BALDRY06'] >= mass_limit

in_sample['in_volume'] = in_volume
in_sample['in_mass'] = in_mass

in_all = (in_sample['in_volume']) & (in_sample['in_mass']) & (in_sample['in_spiral'])
```

# First, compare the gas properties:

```python
sample_a70 = a70_data[in_all]
sample_data = gz_data[in_all]
sample_m = in_sample['m'][in_all]
sample_bar = in_sample['bar'][in_all]
```

```python
from alfalfa_conversions import mass_from_redshift

HI_mass_limit = mass_from_redshift(0.055)
print('{} detections above the completeness limit(>{})'.format(np.sum(sample_a70['loghimass'] >= HI_mass_limit),
                                                               np.round(HI_mass_limit,decimals=2)))

in_reduced = Table()
in_reduced['has_a70'] = sample_a70['loghimass'] >= HI_mass_limit
```

```python
_ = plt.scatter(gz_data['ra_1'],gz_data['dec_1'])
_ = plt.scatter(a70_data['HIra'],a70_data['HIdec'],color='r')
plt.xlabel('RA')
plt.ylabel('DEC')
plt.axis((0,360,-20,90))

in_region = (sample_data['dec_1'] > -0.2) & (sample_data['dec_1'] <= 30.2)
print('< {}/{} within the survey region (~50%)'.format(np.sum(in_region),len(in_region)))
```

```python
# check detection 'rates'?
from bar_plots import comparison_bar_chart

detections = Table()

m_labels = ['m=1','m=2','m=3','m=4','m=5+']
bar_labels = ['no bar','bar']
totals = []

for m, label in enumerate(m_labels):
    m_detection = in_reduced[sample_m == m]['has_a70'] == True
    detections[label] = np.array([np.sum(m_detection)])
    totals.append(len(m_detection)/2) # /2, as survey area is approximately 1/2 of our SDSS data.

for b, label in enumerate(bar_labels):
    bar_detection = in_reduced[sample_bar == b]['has_a70'] == True
    detections[label] = np.array([np.sum(bar_detection)])
    totals.append(len(bar_detection)/2) # divide by 2, as survey area is approximately 1/2 of our SDSS data.
```

```python
labels = [r'Detected in $\alpha 70$']

_ = comparison_bar_chart(detections,labels,'b','percent',linewidth=1,alpha=0.5,
                         ylabel=r'$\%$ detected',column_totals=totals)

plt.vlines(5,0,50,color='k',lw=3)

# more likely to be detected in many-armed or unbarred galaxies.
```

```python
sample_nsa = nsa_data[in_all]
FNugriz_Mags = np.array([sample_nsa['ABSMAG'][:,c] for c in range(7)]).T
FNugriz_errors = np.array([sample_nsa['AMIVAR'][:,c] for c in range(7)]).T
Mags = Table(FNugriz_Mags,names='FNugriz')
Mag_errors = Table(FNugriz_errors,names='FNugriz')

for f,filterband in enumerate('FNugriz'):
    in_reduced['in_'+filterband] = Mag_errors[filterband] > 0

has_FN = (in_reduced['F']) & (in_reduced['N']) & (in_reduced['r'])
in_reduced['FNr'] = has_FN
```

```python
labels = ['Detected in \n GALEX FUV/NUV']

_ = comparison_bar_chart(detections,labels,'red','percent',linewidth=1,alpha=0.5
                         ,ylabel=r'$\%$ detected',column_totals=totals)

plt.vlines(5,0,100,color='k',lw=3)

# more likely to be detected in many-armed or unbarred galaxies.
```

# IR SFR indicator (Chang+15)

```python
sample_chang_out = chang_out[in_all]

in_reduced['has_w3'] = sample_chang_out['flag_w3'] == 1
in_reduced['has_w4'] = sample_chang_out['flag_w4'] == 1
in_reduced['has_wise'] = ((in_reduced['has_w3']) | (in_reduced['has_w4'])) & (sample_chang_out['flag_w'] == 1)
```

```python
detections = Table()

m_labels = ['m=1','m=2','m=3','m=4','m=5+']
bar_labels = ['no bar','bar']
totals = []

for m, label in enumerate(m_labels):
    m_detection = in_reduced[sample_m == m]['has_wise'] == True
    detections[label] = np.array([np.sum(m_detection)])
    totals.append(len(m_detection)) 

for b, label in enumerate(bar_labels):
    bar_detection = in_reduced[sample_bar == b]['has_wise'] == True
    detections[label] = np.array([np.sum(bar_detection)])
    totals.append(len(bar_detection))
```

```python
labels = ['Detected in \n W3 or W4']

_ = comparison_bar_chart(detections,labels,'y','percent',linewidth=1,alpha=0.5
                         ,ylabel=r'$\%$ detected',column_totals=totals)

plt.vlines(5,0,100,color='k',lw=3)

# more likely to be detected in many-armed or unbarred galaxies.
```

```python
from wise_conversions import W3_to_SFR, W4_to_SFR

L_w34 = Table([sample_chang_out['lrest_w3'],sample_chang_out['lrest_w4']])
L_w34['lrest_w3'][in_reduced['has_w3'] == False] = -999
L_w34['lrest_w4'][in_reduced['has_w4'] == False] = -999

sfr_wise = np.ones(len(L_w34))*-999
sfr_wise[in_reduced['has_w4'].astype(bool)] = W4_to_SFR(L_w34['lrest_w4'][in_reduced['has_w4'].astype(bool)])
sfr_wise[in_reduced['has_w3'].astype(bool)] = W3_to_SFR(L_w34['lrest_w3'][in_reduced['has_w3'].astype(bool)])
```

```python
bins = np.linspace(-1.5,1.5,26)

fig,axarr = plt.subplots(5,1,figsize=(10,15),sharex=True,sharey=True)
plt.subplots_adjust(hspace=0,wspace=0)

colors_long = ['orange','r','m','g','b']

for m in range(5):
    plt.sca(axarr[m])
    sfr_m = sfr_wise[sample_m == m]
    
    _ = plt.hist(sfr_wise,color='k',alpha=0.4,histtype='stepfilled',normed=True,bins=bins)
    _ = plt.hist(sfr_m,color=colors_long[m],lw=3,histtype='step',normed=True,bins=bins)
    
    plt.text(0.975,0.925,'{} ({} galaxies)'.format(m_labels[m],np.sum(sfr_m > -99))
             ,transform=axarr[m].transAxes,ha='right',va='top')
    plt.ylabel('normalised \n density')

plt.xlabel('SFR (IR)')
_ = plt.xlim(-1.5,1.5)
```

# Stacked comparison

```python
sfrs = [sfr_mpa,sfr_uv,sfr_wise]
bins=np.linspace(-1.5,1.5,31)

fig_labels = ('MPA', 'FUV', 'WISE')

for m in range(5):
    
    fig, axarr = plt.subplots(3,1,figsize=(7,7),sharex=True,sharey=True)
    plt.subplots_adjust(hspace=0,wspace=0)
    select_m = sample_m == m
    for s in range(3):
        plt.sca(axarr[s])
        sfrs_m = sfrs[s][select_m]
        _ = plt.hist(sfrs[s],color='k',alpha=0.4,histtype='stepfilled'
                     ,bins=bins,normed=True)
        _ = plt.hist(sfrs_m,color=colors_long[m],lw=3,
                     histtype='step',bins=bins,normed=True)
        plt.text(0.975,0.925,fig_labels[s]+ '({})'.format(np.sum(sfrs_m > -999)),
                 transform=axarr[s].transAxes,ha='right',va='top')
    axarr[-1].set_xlabel(r'$SFR \, (M \mathrm{_\odot yr^{-1}})$')
    plt.savefig('output_figures/stacked_sfr_{}'.format(m))
```

```python
masses = sample_data['LOGMSTAR_BALDRY06'] + 0.1 # chabrier
ssfrs = [sfr_mpa-masses,sfr_uv-masses,sfr_wise-masses]
bins=np.linspace(-12,-9,31)

fig_labels = ('MPA', 'FUV', 'WISE')

for m in range(5):
    
    fig, axarr = plt.subplots(3,1,figsize=(7,7),sharex=True,sharey=True)
    plt.subplots_adjust(hspace=0,wspace=0)
    select_m = sample_m == m
    for s in range(3):
        plt.sca(axarr[s])
        ssfrs_m = ssfrs[s][select_m]
        _ = plt.hist(ssfrs[s],color='k',alpha=0.4,histtype='stepfilled'
                     ,bins=bins,normed=True)
        _ = plt.hist(ssfrs_m,color=colors_long[m],lw=3,
                     histtype='step',bins=bins,normed=True)
        plt.text(0.975,0.925,fig_labels[s]+ '({})'.format(np.sum(ssfrs_m > -999)),
                 transform=axarr[s].transAxes,ha='right',va='top')
    axarr[-1].set_xlabel(r'$sSFR \, (yr^{-1})$')
    plt.savefig('output_figures/stacked_ssfr_{}'.format(m))
```
