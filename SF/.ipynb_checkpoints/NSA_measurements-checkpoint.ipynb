```python
# Preamble

from astropy.table import Table, column
import numpy as np
import os

%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
# better-looking plots
plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.figsize'] = (10.0, 8)
plt.rcParams['font.size'] = 18
mpl.ticker.AutoLocator.default_params['nbins'] = 5
mpl.ticker.AutoLocator.default_params['prune'] = 'both'

mpl.rcParams['ps.useafm'] = True
mpl.rcParams['pdf.use14corefonts'] = True
mpl.rcParams['text.usetex'] = True
```

```python
import sys
sys.path.append('../../astro_codes/')
```

```python
samples = Table.read('matched_fits/samples.fits')
nsa_data = Table.read('matched_fits/nsa_matched.fits')
s16_data = Table.read('matched_fits/s16_matched.fits')
```

```python
hdelta = nsa_data['AHDEW']
hdelta_error = nsa_data['AHDEWERR']
has_hdelta = hdelta_error > 0

d4000 = nsa_data['D4000']
d4000_error = nsa_data['D4000ERR']
has_d4000 = d4000_error > 0

a_fraction = nsa_data['FA']
a_fraction_error = nsa_data['FAERR']
has_a_fraction = a_fraction_error > 0

b300 = nsa_data['B300']
b1000 = nsa_data['B1000']
has_b1000 = nsa_data['B300'] > 0
has_b300 = nsa_data['B1000'] > 0

in_mass_limit = (samples['in_mass_limit']) & (has_d4000) & (has_hdelta)
spiral_ok = (samples['in_spiral']) & (samples['axial_ok'])
```

```python
afuv = s16_data['afuv']
has_afuv = s16_data['afuv'] > 0

av = s16_data['av']
has_av = s16_data['av'] > 0
```

```python
def plot_xy(x,y,in_sample,in_spiral,m_values,xylims=None,xlabel='x',ylabel='y'):

    fig, axarr = plt.subplots(2,3,figsize=(20,13.4),sharex=True,sharey=True)
    plt.subplots_adjust(hspace=0,wspace=0)
    axarr = axarr.ravel()
    fig.delaxes(axarr[-1])
    colors = ['orange','r','m','g','b']

    for m in range(5):
        spirals_m = (in_spiral) & (m_values == m) & (in_sample)
        plt.sca(axarr[m])
        _ = plt.scatter(x[in_sample],y[in_sample],color='k',alpha=0.1)
        _ = plt.scatter(x[spirals_m],y[spirals_m],color=colors[m],alpha=0.5)
    
    axarr[4].set_xlabel(xlabel)
    axarr[0].set_ylabel(ylabel)
    axarr[3].set_ylabel(ylabel)
    if xylims is not None:
        axarr[0].axis(xylims)
    
    return None

in_sample = (has_d4000) & (has_hdelta) & (in_mass_limit)
in_spiral = (in_sample) & (spiral_ok)
plot_xy(d4000,hdelta,in_sample,in_spiral,samples['m'],(1,2.5,-5,5),'D4000','H$\delta$')

in_sample = (has_d4000) & (has_a_fraction) & (in_mass_limit)
in_spiral = (in_sample) & (spiral_ok)
plot_xy(d4000,a_fraction,in_sample,in_spiral,samples['m'],(1,2.5,0,0.6),'D4000','$f$(A)')

in_sample = (has_d4000) & (has_b300) & (in_mass_limit)
in_spiral = (in_sample) & (spiral_ok)
plot_xy(d4000,b300,in_sample,in_spiral,samples['m'],(1,2.5,0,0.1),'D4000','b300')

in_sample = (has_d4000) & (has_b1000) & (in_mass_limit)
in_spiral = (in_sample) & (spiral_ok)
plot_xy(d4000,b1000,in_sample,in_spiral,samples['m'],(1,2.5,0,0.8),'D4000','b1000')
```

```python
def plot_hists(x,bins,in_sample,in_spiral,m_values,xlabel='x'):

    colors = ['orange','r','m','g','b']
    
    fig, axarr = plt.subplots(5,1,figsize=(10,15),
                              sharex=True,sharey=True)

    plt.subplots_adjust(hspace=0,wspace=0)

    for m in range(5):
        plt.sca(axarr[m])
    
        spirals_m = (in_spiral) & (m_values == m) & (in_sample)
        _ = plt.hist(x[in_sample],bins=bins,normed=True,
                     alpha=0.4,color='k',histtype='stepfilled')
        _ = plt.hist(x[spirals_m],bins=bins,normed=True,
                     linewidth=3,color=colors[m],histtype='step')
        plt.ylabel('nomralised \n density')

    axarr[-1].set_xlabel(xlabel)
    return None

'''
in_sample = (in_mass_limit) & (has_d4000)
in_spiral = (in_sample) & (spiral_ok)
plot_hists(d4000,np.linspace(1,2.2,25),in_sample,in_spiral,samples['m'],'D4000')

in_sample = (in_mass_limit) & (has_a_fraction)
in_spiral = (in_sample) & (spiral_ok)
plot_hists(a_fraction,np.linspace(0,0.5,21),in_sample,in_spiral,samples['m'],'$f$(A)')

in_sample = (in_mass_limit) & (has_b300)
in_spiral = (in_sample) & (spiral_ok)
plot_hists(b300,np.linspace(0,0.04,17),in_sample,in_spiral,samples['m'],'$b_{300}$')

in_sample = (in_mass_limit) & (has_b1000)
in_spiral = (in_sample) & (spiral_ok)
plot_hists(b1000,np.linspace(0,0.4,17),in_sample,in_spiral,samples['m'],'$b_{1000}$')
'''

in_sample = (in_mass_limit) & (has_afuv)
in_spiral = (in_sample) & (spiral_ok)
plot_hists(afuv,np.linspace(0,5,51),in_sample,in_spiral,samples['m'],'$A_{FUV}$')

in_sample = (in_mass_limit) & (has_av)
in_spiral = (in_sample) & (spiral_ok)
plot_hists(afuv,np.linspace(0,5,51),in_sample,in_spiral,samples['m'],'$A_{V}$')
```

```python
from compare_xy import compare_xy_data

colors = ['orange','r','m','g','b']

x_all = afuv
y_all = av
in_ml = samples['in_mass_limit']
in_sample = ((afuv > 0) & (av > 0))

in_reference = (in_ml) & (in_sample) & (samples['in_spiral']) & (samples['axial_ok'])
x_reference, y_reference = x_all[in_reference], y_all[in_reference]

xy_range = ((1,5,0,1))
'''
fig, axarr = plt.subplots(2,3,figsize=(20,13.4),sharex=True,sharey=True)
plt.subplots_adjust(hspace=0,wspace=0)
axarr = axarr.ravel()
fig.delaxes(axarr[-1])
'''

refererence_hist_fill_properties = {'alpha':0.4}

reference_contour_fill_properties = {'alpha':1}

#refe
bandwidth = None
for m in range(5):
    
    m_ok = samples['m'] == m
    in_m = ((m_ok) & (in_sample) & (in_ml) &
            (samples['in_spiral']) & (samples['axial_ok']))
    
    x, y = x_all[in_m], y_all[in_m]
    
    fig, axarr, bandwidth = compare_xy_data(x_reference,y_reference,xy_range,N_bins=(21,21),
                                            bandwidth=bandwidth,fill=True,
                                            hist_fill_properties=refererence_hist_fill_properties,
                                            contour_fill_properties=reference_contour_fill_properties)
    
    line_properties = {'color':colors[m],
                       'linewidth':2}
    
    fig, axarr, _ = compare_xy_data(x,y,xy_range,N_bins=(21,21),
                                   line_properties=line_properties,
                                   fig=fig,axarr=axarr)
    
    axarr[2].set_xlabel('$A_{FUV}$')
    axarr[2].set_ylabel('$A_{V}$')
```
