```python
def get_assignments(data,question_dictionary,th=None):

    answers = question_dictionary['t11_arms_number']['answers']
    
    arm_answers = ['t11_arms_number_' + A + '_debiased_rh' for A in answers[:-1]]
    arm_array = np.array([data[A] for A in arm_answers]).T
    raw_answers = ['t11_arms_number_' + A + '_weighted_fraction' for A in answers[:-1]]
    raw_array = np.array([data[A] for A in raw_answers]).T
    
    if th != None:
        select = np.max(arm_array,axis=1) > th
    else:
        select = np.ones(len(data)) == 1
    
    arm_assignments = np.argmax(arm_array,axis=1)
    raw_assignments = np.argmax(raw_array,axis=1)
    
    arm_assignments = arm_assignments[select]
    raw_assignments = raw_assignments[select]
    
    return arm_assignments,raw_assignments


def reduce_sample(full_data,questions,question,p_cut=0.5,N_cut=5,normalised_values=True):
    
    # Get the reference sample from the previous data:
    
    previous_q = questions[question]['pre_questions']
    previous_a = questions[question]['pre_answers']
    
    if normalised_values == True:
        suffix = '_debiased_rh_normalised'
    else:
        suffix = '_debiased_rh'
    
    if previous_q != None:
        
        p_col = np.ones(len(full_data))
        
        for m in range(len(previous_q)):
            p_col = p_col*(full_data[previous_q[m] + '_' + previous_a[m] + suffix])
        N_col = (full_data[previous_q[-1] + '_' + previous_a[-1] + '_count'])
        
        select = (p_col > p_cut) & (N_col >= N_cut)
        data_reduced = full_data[select]
        print('{}/{} galaxies with p>{} and N>={}.'.format(len(data_reduced),
                                                          len(full_data),p_cut,N_cut))
    
    else:
        data_reduced = full_data.copy()
        select = np.ones(len(full_data)) == 1
        print('Primary question, so all {} galaxies used.'.format(len(data_reduced)))
    
    return select
```

```python
sfrs = Table.read('../fits/MPA_JHU/gal_totspecsfr_dr7_v5_2.fits')
ids = Table.read('../fits/MPA_JHU/CAS-IDs.fits')
bpt_classes = Table.read('../fits/MPA_JHU/gal_iclass_table_dr7_v5_2.fits')
sfrs['class'] = bpt_classes['I_CLASS']

galaxy_data = Table.read('../fits/full_sample_debiased.fits')
```

```python
os.chdir('../Debiasing/')

import make_dictionaries
questions = make_dictionaries.questions

os.chdir('../star_formation/')
```

```python
# match the galaxies by DR7 ID?
matches = Table.read('../fits/MPA_JHU/mpa_matches.fits')

print('{} matches found!'.format(len(matches)))

matched_galaxies = galaxy_data[matches['i_galaxies']]
matched_mpa = sfrs[matches['i_mpa']]
matched_ids = ids[matches['i_mpa']]
```

```python
# assign arm number/ get spirals:

assignments,_ = get_assignments(matched_galaxies,questions)
in_volume_limit = matched_galaxies['in_volume_limit'] == True
in_mass_limit = matched_galaxies['LOGMSTAR_BALDRY06'] > 10.6
spiral = reduce_sample(matched_galaxies,questions,'t11_arms_number')
remove_ct = (matched_galaxies['t04_spiral_a08_spiral_count'] 
             - matched_galaxies['t11_arms_number_a37_cant_tell_count'] >= 5)

select_sample = (spiral) & (remove_ct) & (in_volume_limit) & (in_mass_limit)
```

```python
gal_sp = matched_galaxies[select_sample]
mpa_sp = matched_mpa[select_sample]
assign_sp = assignments[select_sample]

sf = (mpa_sp['class'] == 1) & (mpa_sp['FLAG'] == 0)
low_sn_sf = (mpa_sp['class'] == 2) & (mpa_sp['FLAG'] == 0)
print('{} SF galaxies; {} are low S/N SF ({} total)'.format(np.sum(sf),np.sum(low_sn_sf),len(mpa_sp)))
```

```python
fig,axarr = plt.subplots(5,1,figsize=(10,15),sharex=True,sharey=True)
plt.subplots_adjust(hspace=0,wspace=0)

colors_long = ['orange','red','m','g','b','k']

for m in range(5):
    
    sel_m = mpa_sp[((sf) + (low_sn_sf)) & (assign_sp == m)]
    
    _ = axarr[m].hist(sel_m['class'],bins=[0.75,1.25,1.75,2.25],normed=True
                      ,histtype='stepfilled',alpha=0.8,color=colors_long[m])
    axarr[m].set_ylabel('f')
    
axarr[-1].set_xlabel('SF --$>$ none-SF')
```
